### ## Bivariate Analysis: Book Quality vs Popularity
### 
### **Research Question:** Do higher-rated books receive more ratings (popularity)?
### 
### **Variables:**
### - X-axis: Book_average_rating (quality indicator, scale ~3.0-5.0)
### - Y-axis: Book_ratings_count (engagement/popularity indicator)
### 
### **Why scatter plot?** Reveals correlation patterns and density clusters

```python
plt.scatter(df['Book_average_rating'],df["Book_ratings_count"])
plt.xlabel("Book average pricing")
plt.ylabel("Book Rating Count")
plt.title("Scatterplot of book average rating vs book rating count")
plt.show()
```

### ### Key Observations:
### 
### | Pattern | Finding |
### |---------|---------|
### | **Central Cluster** | Most books cluster around 3.75-4.25 rating with 75K-150K ratings |
### | **Weak Positive Trend** | Slight upward slope: higher ratings → more ratings |
### | **High-Rating Outliers** | Some 4.5+ rated books reach 200K+ ratings |
### | **Low-Rating Survivors** | Even 3.0-3.5 rated books maintain 50K+ ratings |
### | **Ceiling Effect** | Maximum popularity (~200K) achieved across rating range 3.75-4.5 |
### 
### ### Critical Insights:
### 
### 1. **Quality ≠ Guaranteed Popularity**
###    - Books with 3.5 ratings can be as popular as 4.5-rated books
###    - Suggests marketing/genre factors may outweigh pure quality
### 
### 2. **Sweet Spot Exists**
###    - Ratings between 3.8-4.2 show highest density of popular books
###    - "Good enough" quality + broad appeal = maximum engagement
### 
### 3. **Diminishing Returns**
###    - Above 4.5 rating, no significant increase in rating count
###    - Niche excellence doesn't translate to mass popularity
### 
### ### Note on Axis Label:
### X-axis label says "pricing" but appears to show "rating" values (3.0-4.75).
### Verify if this is a typo in `plt.xlabel()` - should likely be "Book Average Rating".
### 
### ### Business Implication:
### Focus on achieving 4.0+ rating threshold, but invest equally in marketing 
### and genre positioning - quality alone doesn't drive popularity.
